STEMMING

is the proces of reducing a wod which word from root word

```python
# A pocess of reducing a word that has a root word example acting,actress to act
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content) # pass text column to remove everything tthat is not in Upper or lower case (Alphabets)
  stemmed_content = stemmed_content.lower() # making our content be in lower case
  stemmed_content = stemmed_content.split() # split all words and put it in a list
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')] # removing all words that belong to stop words
  stemmed_content = ' '.join(stemmed_content) # join all the words in the list
  return stemmed_content
```

```python
twitter_data['stemmed_content']=twitter_data['text'].apply(stemming) # applying stemming
```

```python
#separating the data and label
X=twitter_data['stemmed_content'].values
Y=twitter_data['target'].values
#
```

Splitting the  data to training data and test data

```python
X_train,X_test,Y_train,Y_test= train_test_split (X,Y, test_size=0.2 ,starify=Y,random_state=2)
```

```python
print(X.shape X_train.shape,X_test.shape)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
# converting the textual data to numerical data
vectorizer=TfidfVectorizer()

X_train =vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

TRAINING THE MACHINE LEARNING MODEL

```python
model=LogisticRegression(max_iter=1000) # giving a variable name
```

```python
model.fit(X_train,Y_train)
```

MODEL EVALUATION

Accuracy

```python
#Accuracy score on the training data
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy score prediction:", training_data_accuracy)
```

```python
# we now check the score on the test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score prediction on the training data:", test_data_accuracy)
```
